```
find the best price that optimizes the profit for pineapple juice knowing that the cost of one pineapple juice is 70 cents.
remember that profit is revenue - cost.

```

````python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
````

````python
df=pd.read_csv('Pinapple_juice.csv')
df.head()
````

````python
x=df[['Price']]
y=df['Pinapple juice']
````

````python
model=LinearRegression()
model.fit(x,y)
````

````python
model.intercept_
````

````python
intercept=model.intercept_
coef=model.coef_[0]
````

````python
df.describe()
````

````python
simulation_data=pd.DataFrame({ 'price': np.linspace(0.1,5,300)})
````

````python
simulation_data['demand']= simulation_data['price']*coef + intercept
````

````python
simulation_data
````

````python
simulation_data['cost']= simulation_data['price']*0.7
````

````python
simulation_data
````

````python

simulation_data['revenue']= simulation_data['demand']*simulation_data['price']

````

````python
simulation_data['total_cost']= simulation_data['cost']*simulation_data['demand']
````

````python
simulation_data['profit']=simulation_data['revenue']-simulation_data['total_cost']
````

````python
plt.plot(simulation_data['price'],simulation_data['revenue'],label='revenue')
plt.plot(simulation_data['price'],simulation_data['profit'],label='profit')
plt.plot(simulation_data['price'],simulation_data['demand'],label='demand')
plt.legend()
plt.title('on X axis : Price')
plt.show()
````
