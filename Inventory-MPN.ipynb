```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./retail_clean.csv')
df.info()
```

```python
df=df.drop_duplicates()
df=df.dropna(axis=0)
```

```python
df.shape
```

```python
import datetime
df['date']=pd.to_datetime(df['InvoiceDate']).dt.strftime('%Y-%m-%d')
df['date']=pd.to_datetime(df['date'])
```

```python
df['year']=df['date'].dt.year
```

```python
years_2=df[df['year'].isin([2010,2011])]
```

```python

total=years_2.groupby(['year','Description']).agg(total_sales=('Quantity','sum'),price=('Price','mean')).reset_index()
```

```python
expected=total.groupby('Description').agg(expected_demand=('total_sales','mean'),sd=('total_sales','std'),price=('price','mean'))
```

```python
expected=expected.reset_index()
```

```python
expected
```

```python
import numpy as np
```

```python
def margin_error(df):
    if( pd.isna(df['sd'])):
        a=df['expected_demand']*0.1
    else:
        a=df['sd']
    return a
```

```python
expected['sd1']=expected.apply(margin_error,axis=1)
```

```python
expected['cost']=expected['price']*0.6
```
