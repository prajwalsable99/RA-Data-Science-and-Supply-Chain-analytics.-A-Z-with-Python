  # â€¢ Drop duplicates if any from the dataset.

````python
df['InvoiceDate']= pd.to_datetime( df['InvoiceDate'])
````

# â€¢ get the week , to get the day of the week, the month and the year from invoice date column.

````python
df['week']=df['InvoiceDate'].dt.isocalendar().week
df['dayofweek']=df['InvoiceDate'].dt.dayofweek
df['month']=df['InvoiceDate'].dt.month
df['year']=df['InvoiceDate'].dt.year
````

````python
for col in ['week','dayofweek','month','year']:
    print(df[col].unique())
````

# â€¢ Make a new column and called it month year with the month name and year.

````python
df['month_year']=df['InvoiceDate'].dt.strftime('%B-%Y')
````

````python
df
````

#  Get the last purchase date per customer

````python
df['date']=df['InvoiceDate'].dt.strftime('%Y-%m-%d')
````

````python
df['date']= pd.to_datetime( df['date'])
````

````python
df['date'].head()
````

````python
max_date=df['date'].max()
````

````python
max_date
````

````python
customer_last= df.groupby('Customer ID').agg(last_purchase_date= ('date','max')).reset_index()
````

   
# â€¢get the recency per customer


````python
customer_last['recency']=max_date- customer_last['last_purchase_date']
````

````python
customer_last
````

````python
customer_last['recency']=customer_last['recency'].astype('string').str.replace('days','')
````

````python
customer_last['recency']=customer_last['recency'].astype('int')
````

````python
customer_last.info()
````

````python
plt.hist(customer_last['recency'])
````

#    â€¢ apply two weeks and one week moving average  for sales on the data .


````python
sales_per_day= df.groupby('date').agg(total_sales= ('Quantity','sum'))
````

````python
sales_per_day
````

````python
sales_per_day.plot()
````

````python
sales_per_day['mov_avg_7']=sales_per_day.total_sales.rolling(window=7).mean()
sales_per_day['mov_avg_14']=sales_per_day.total_sales.rolling(window=14).mean()
````

````python
sales_per_day.plot()
````

#    â€¢ resample the data to weekly data using the sum of all observations on that week.



````python
sales_weekly=sales_per_day.total_sales.resample('W').sum()
````

````python
sales_weekly.plot()
````
