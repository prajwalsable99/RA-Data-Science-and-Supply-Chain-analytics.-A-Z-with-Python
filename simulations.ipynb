```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lambda1=1
mean_service=0.8
sd=0.2
```

```python
arrival_times=np.random.exponential(1,400).cumsum()
service_times=np.random.normal(mean_service,sd,400)
```

```python
plt.plot(arrival_times,label='arrival time')
plt.plot(service_times,label='service time')
plt.legend()
plt.show()
```

```python
def avg_waiting_time(arrival_times,service_times ):
    waiting_times=[]
    leaving_times=[]
    waiting_times.append(0)
    leaving_times.append(arrival_times[0]+waiting_times[0]+service_times[0])

    for i in range(1,len(arrival_times)):
        waiting_times.append( max(0,leaving_times[i-1] -arrival_times[i] ))
        leaving_times.append(arrival_times[i]+waiting_times[i]+service_times[i])

    return np.mean(waiting_times)
    
```

```python
avg_waiting_time(arrival_times,service_times)
```
